# Data

```python
df = pd.DataFrame({'fruit':['apple','apple','pear','grapes','grapes'],
                   'color':['green','green','yellow','red','white'],
                   'sales':[13,6,16,8,4],
                   'profit':[1,1,1,1,1]})

df
```

# How to organize a dataframe by specific columns?

```python
pd.pivot_table(df,index=['color'],columns=['fruit'])
```

```python
pd.pivot_table(df,index=['fruit'],columns=['color'])
```

```python
pd.pivot_table(df,values='profit',index=['color'],columns=['fruit'])
```

# How to fill NaN values?

```python
pd.pivot_table(df,index=['color'],columns=['fruit'], fill_value=0)
```

# How to add sub-totals to the columns and rows? 

> Note that the default aggfunc is the average

```python
pd.pivot_table(df,index=['color'],columns=['fruit'], fill_value=0, margins=True)
```

### aggfunc examples:  

* 'sum'  
* 'count'

* 'max'  
* 'min'  
* len  
* aggfunc={'column_name': 'mean'}  
* aggfunc={'column_name':  lambda x: x.sum()}

# Use the ***sum*** function

```python
pd.pivot_table(df,index=['color'],columns=['fruit'], fill_value=0, aggfunc='sum')
```

# Use the ***count*** function

```python
pd.pivot_table(df,index=['color'],columns=['fruit'], fill_value=0, aggfunc='count')
```

# Use the ***mean*** function

```python
pd.pivot_table(df,index=['color'],columns=['fruit'], fill_value=0, aggfunc='mean')
```

# Use the ***max*** function

```python
pd.pivot_table(df,index=['color'],columns=['fruit'], fill_value=0, aggfunc='max')
```

```python
pd.pivot_table(df,index=['color'],columns=['fruit'], fill_value=0, aggfunc=len)
```

# How to apply different functions to different columns?

```python
pd.pivot_table(df,index=['color'],columns=['fruit'],aggfunc={'profit':'max','sales':'mean'}, fill_value=0)
```

# How to apply multiple functions to one column?

```python
pd.pivot_table(df,index=['color'],columns=['fruit'],aggfunc={'sales':['max','min','mean'],'profit':'mean'}, fill_value=0)
```

# How do you apply a custom function?

```python
pd.pivot_table(df,index=['color'],columns=['fruit'],aggfunc={'sales': lambda x: x.sum() if x.sum() > 4 else None})
```

```python
def f1(group):
    """ flip sign of any value grater than 10 """
    
    print group
    r = group.apply(lambda x: -1*x if x > 10 else x)
    return r.sum()

pd.pivot_table(df,index=['color'],columns=['fruit'],aggfunc={'sales': f1})
```

# How to glue pivot tables?

```python
p1 = pd.pivot_table(df,index=['color'],columns=['fruit'], fill_value=0, aggfunc=len)
p2 = pd.pivot_table(df,index=['color'],columns=['fruit'], fill_value=0, aggfunc='max')

# rename levels
p2 = p2.rename_axis({'profit':'profit_max', 'sales':'sales_max'}, axis=1)

pd.concat([p1,p2], axis=1)
```
