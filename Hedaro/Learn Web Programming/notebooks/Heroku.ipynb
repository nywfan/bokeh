# How to get you Python web app on Heroku?  

***Step 1:*** Go to https://www.heroku.com/ and sign up for an account. All you need to do is provide your email address, create a password and that's pretty much it.  

you will soon get an email stating:  

> Thanks for signing up with Heroku! You must follow this link to activate your account  

Click on the link and you are ready to go.  

***Step 2:*** Download the [Heroku toolbelt](https://toolbelt.heroku.com/) and do a full install. This will install three seperate programs.  

* Heroku - command line tool for deploying and managing your web apps  
* Git - for creating and managing your git repositories  
* Ruby - mainly for running a local web server  

***Windows users***  
I had a truck load of issues trying to get everything working after the install. Here are a couple of pointers to getting this entore setup working correctly on windows.  

* Please install in a directory with no spaces. For example, C:\\Heroku  
* After installation you have to manually add the ruby reference to your system variables. 

> Go to: Control Panel\\System and Securit\\System  
> Click on advanced systems settings  
> Click on environment variables 
> Under system variables, edit path variable

You should see the following. If not, please add and hit OK.

> C:\\Heroku\\bin;C:\\Program Files (x86)\\git\\cmd;C:\\Heroku\\ruby-1.9.2\\bin

* on the git bash terminal, delete and install a different [version of foreman](http://stackoverflow.com/questions/19078939/foreman-installed-by-heroku-toolbelt-on-windows-cant-be-found)  

> gem uninstall foreman  
> gem install foreman -v 0.61  



# Lets now set up our web app for deployment  

Make a new folder and stick the contents of ***"01 - Lesson"*** inside the folder.  

* static - this folder should be empty  
* templates - contains index.HTML and template.HTML  
* app.py  

***step 1:*** were going to have to modify the app.py and change the way we set our port.  

    # Webserver stuff
    import tornado.ioloop
    import tornado.web
    from tornado.options import define, options
    
    # route to index.html
    class MainHandler(tornado.web.RequestHandler):
        def get(self):
            self.render('templates/index.html') 
    
    # r"/" == root website address
    application = tornado.web.Application([
        (r"/", MainHandler)
    ],debug=True) 
    
    define("port", default=8888, help="run on the given port", type=int)   
    
    # Start the server at port n
    if __name__ == "__main__":
        tornado.options.parse_command_line()
        application.listen(options.port)
        tornado.ioloop.IOLoop.instance().start()
    

***The new lines are:***  
    from tornado.options import define, options  
    define("port", default=8888, help="run on the given port", type=int)  
    application.listen(options.port)  
        
        

***step 1:*** add a text file named runtime.txt and write the line below. This file tells Heroku what version of Python we want to use.    

> python-2.7.4  

***step 2:*** add a text file named requirements.txt and write the line below. This file tells Heroku that we are running a Python web app and tells Heroku what libraries need to be installed. Heroku then will pip install all of these upon deployment.    

> tornado    

***step 3:*** add a file named Procfile (remove any file extension) and write the line below. This file tells Heroku what file to run initially.  

> web: python app.py --port=$PORT  





# Create a git repository and push our web app to Heroku  

***step 1:*** Open git bash and navigate to the folder of your web app.

> i.e. cd c:\\yourwebapp  

***step 2:*** create a repository    

> git init  
> git add -A  
> git commit -m "created repo" 

***step 3:*** create a Heroku repo    

> heroku login  
> heroku create  

lets rename the project. Please use a different name.  

> heroku rename hdrojas  
> git push heroku master  

***step 4:*** Open your web app in your favorite browser  

> http://yourappname.herokuapp.com/
